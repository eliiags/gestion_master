```python
# 7. Se pide una palabra hasta introducir una palabra de mínimo 5 letras cuya primera letra sea la c o C
while (True):
    print("Introduce una palabra : ", end="")
    palabra = input()
    if (len(palabra) == 5 and (palabra.find("c")==0 or palabra.find("C")==0)):
        break
    else:
        continue
```

```python
# Avanzado, solo en aquellos casos que se superen todos los ejercicios de este tema con mucha soltura:
# Juego Adivina Palabra.
# Primer jugador introduce una palabra por teclado.
# Segundo jugador introduce 10 letras y el script nos dice una a una si la palabra contiene esa letra.
# Una vez introducida las 10 letras el script dirá el numero de aciertos y fallos.
# El segundo jugador introduce la palabra que cree que es.
# El script informa si ha ganado o no.
print("Primer jugador: Introduce una palabra: ", end="")
palabra = input()
print("Segundo jugador: Introduce 10 letras consecutivas: ", end="")
letras = input()

palabra.lower()
letras.lower()
i = 0
aciertos = ""
a = 0

while(i < len(letras)):
    if (palabra.find(letras[i]) != -1):
        aciertos += letras[i] + " "
        a+=1
    i=i+1

print("Has acertado en las letras : " + aciertos)
print("Has tenido " + str(a) + " aciertos y " + str(10-a) + " fallos")
print("¿Cual es la palabra?")
ganador = input()

if (ganador == palabra):
    print("HAS GANADO!!")
else:
    print("has perdido :(")
```
